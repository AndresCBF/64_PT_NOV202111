```python
lista = [1,[1,2, [3, 4, 5]], [2,3, [4, 5, 6]]]
for item in lista:
    if type(item) != list:
        print(item)
    new_item = item
    while type(new_item) == list:
        current_item = new_item
        
        for newer_item in new_item:
            if type(newer_item) == list:
                new_item = newer_item
            else:
                print(newer_item)
        
        if current_item == new_item:
            break
```
